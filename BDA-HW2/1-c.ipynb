```python
import numpy as np
import scipy.stats
from matplotlib import pyplot as plt
```

```python
def draw_normal(mu=0, sigma=1, size=10000):
	np.random.seed(37710)
	normal = np.random.normal(mu, sigma, size)
	_, bins_edge, _ = plt.hist(normal, bins=50, density=True, alpha=0.5)

	y = scipy.stats.norm.pdf(bins_edge, mu, sigma)
	plt.plot(bins_edge, y, label='$\mu$=%.1f, $\sigma^2$=%.1f' % (mu, sigma))
	plt.xlim(-1, 3)

def show_normal_dist_plot():
	draw_normal(mu=1, sigma=0.1, size=10000)
	draw_normal(mu=1, sigma=2, size=10000)
	draw_normal(mu=1, sigma=5, size=10000)
	plt.legend(loc=0, ncol=1)
	plt.show()
```

```python
show_normal_dist_plot()
```
