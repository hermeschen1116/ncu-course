```python
import pymc as pm
import arviz as az
```

```python
with pm.Model():
    normal_distribution1 = pm.Normal("N(1,4)", mu=1, sigma=2)
    normal_distribution2 = pm.Normal("N(2,4)", mu=2, sigma=2)
    samples = pm.sample(10000, random_seed=37710)
```

```python
az.plot_posterior(samples["posterior"]["N(1,4)"] * 0.5 + samples["posterior"]["N(2,4)"] * 0.5)
```
